```python
# The csv file contains 151 seconds of walking at 5 km/h, recorded at the right hand. 
import pandas as pd
walkDf = pd.read_csv('walking.csv', 
                 parse_dates=True, 
                 skiprows=0, 
                 index_col='time', 
                 header=0, 
                 names=['time','ax','ay','az','|a|']) 
walkDf.index = pd.to_datetime(walkDf.index, format='%H:%M:%S:%f')
walkDf.head()
```

```python
# Task 3
# The device was supposed to record at 100Hz. Visualising the first and last 3000 samples should give us around 30s of data.
# This will easily allow us to see the transitions at the start and end of recording
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def visualiseData(dataset, activity):
    timeInSeconds = dataset.index - dataset.index[0]
    
    timeScale = timeInSeconds[0:3000].seconds + timeInSeconds[0:3000].microseconds/1000000 # observe the start of series
    data1 = dataset['|a|'].iloc[0:3000]
    
    timeScale2 = timeInSeconds[-3000:-1].seconds + timeInSeconds[-3000:-1].microseconds/1000000 # "     "  end of series
    data4 = dataset['|a|'].iloc[-3000:-1]

    fig, axs = plt.subplots(1, 2, figsize=(20,7)) 
    fig.suptitle('Beginning and end of signal ('+activity+')', y=1.02, fontsize='xx-large')
    axs[0].plot(timeScale, data1)
    axs[0].set_title('|a| start', fontsize='x-large')
    axs[0].set_xlabel('Time (s)', fontsize='x-large')
    axs[0].set_ylabel('Acceleration (m/s^2)', fontsize='x-large')
    
    axs[1].plot(timeScale2, data4)
    axs[1].set_title('|a| end', fontsize='x-large')
    axs[1].set_xlabel('Time (s)', fontsize='x-large')
    
    fig.tight_layout(pad=1.0)
    # Repeated axes labels (i.e. acceleration) have been removed to improve readability
```

```python
import matplotlib
matplotlib.get_backend()
```

```python
visualiseData(walkDf, 'walking')
```

There appears to be a transition period of around 3 seconds at the start and end of the signal  
The range of acceleration is about (0, 8) m/s^2. 

```python
visualiseData(runDf, 'running')
```

There appears to be a transition period of around 3 seconds at the start and 5 seconds at the end of the signal.  
The range of acceleration is about (0, 50) m/s^2 respectively.  
This is quite a lot higher than the walking signals, which is to be expected.  
Each cycle is also a lot shorter, but it is difficult to quantify the change in period from these plots alone.

```python
# Task 4
# Remove the first and last 3 seconds of each signal to remove the transition periods
import datetime

startTime = walkDf.index[0] + pd.DateOffset(seconds=3)
endTime = walkDf.index[-1] + pd.DateOffset(seconds=-3)
walkDf = walkDf[(walkDf.index > startTime) & (walkDf.index < endTime)]
walkDf.head()

# Remove the last 10 seconds for running
startTime = runDf.index[0] + pd.DateOffset(seconds=3)
endTime = runDf.index[-1] + pd.DateOffset(seconds=-5)
runDf = runDf[(runDf.index > startTime) & (runDf.index < endTime)]

walkDf['delta_t'] = (walkDf.index - walkDf.index[0]).fillna(0)
runDf['delta_t'] = (runDf.index - runDf.index[0]).fillna(0)
# note that the change in time is varying
```

```python
# Visualising the data after pruning
visualiseData(walkDf, 'walking')
```

```python
visualiseData(runDf, 'running')
```

```python
print('walking dataset duration = ' + str(walkDf.index[-1] - walkDf.index[0]))
print('running dataset duration = ' + str(runDf.index[-1] - runDf.index[0]))
print('number of walking dataset samples = ' + str(len(walkDf)) )
print('number of running dataset samples = ' + str(len(runDf)) )
```

```python
# Task 5
# remove incomplete time period
incompleteWindow = len(walkDf)%100
walkDf = walkDf.iloc[:-incompleteWindow]
walkDf2 = walkDf.reset_index()
idx = np.asarray(np.floor(walkDf2.index/100).values, np.int)
walkDf['window_idx'] = ['w{:03}'.format(i) for i in idx]
walkDf
```

```python
# remove incomplete time period

incompleteWindow = len(runDf)%100
runDf = runDf.iloc[:-7]
runDf2 = runDf.reset_index()
idx = np.asarray(np.floor(runDf2.index/100).values, np.int)
runDf['window_idx'] = ['r{:03}'.format(i) for i in idx]
runDf
```

```python
df = pd.concat([walkDf, runDf], ignore_index=False)
df
```

```python
print('There are ' + str(df.shape[0]/100) + ' unique windows')
```

```python
from tsfresh.feature_extraction.settings import ComprehensiveFCParameters
fc_parameters = ComprehensiveFCParameters()
```

```python
from tsfresh.feature_extraction import extract_features

X = extract_features(df, 
                     column_id='window_idx', 
                     column_sort='delta_t',
                    default_fc_parameters = fc_parameters
                    )
```

```python
# save the original feature matrix for debugging
saveX = X
```

```python
print('The dimensions of the feature matrix are ' + str(X.shape))
print('There are ' + str(X.shape[1]) + ' unique features')
```

```python
# remove invalid features 
X = X.dropna(axis=1)
X
```

```python
X.columns.values
```

```python
X.shape
```

```python
# Task 6
yDf = X.index.str.startswith('r')
y = pd.Series(yDf, index=X.index)
y
```

```python
y.shape
```

```python
from tsfresh.transformers import FeatureSelector
select = FeatureSelector()
select.fit(X, y)
```

```python
select.features
```

```python
p_values = pd.Series(select.p_values, index=select.features)
p_values
```

```python
p_values[select.relevant_features]
# The first 5 are the most relevant features/ smallest p-values
```

```python
print('The 5 most significant features are:\n')
for i in range(0,5):
    print(select.relevant_features[i])
```

```python
import seaborn as sns # visualising the features

feature_matrix_data = {'isRunning': yDf}
for i in range(0,5):
    feature_matrix_data[select.relevant_features[i]] = X[select.relevant_features[i]]

    
feature_matrix = pd.DataFrame(data=feature_matrix_data)
feature_matrix.head
```

```python
sns.pairplot(feature_matrix, 
             hue='isRunning',
             corner=True,
             diag_kind='hist'
            )
```

The individual (diagonal) features tend to be bimodal, except for az__variance_larger_than_standard_deviation.
ay__range_count__max_1__min_-1 and ax__range_count__max_1__min_-1 form two linearly seperable clusters.  
Most feature-pairs are entirely seperable.  
It is interesting that some of the walking data had az_variance_greater_than_standard_deviation == 1 (2nd diagonal plot)

```python
# Apparently feature scaling is not required for Random Forests -
# https://stackoverflow.com/questions/8961586/do-i-need-to-normalize-or-scale-data-for-randomforest-r-package
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_features=5)
clf.fit(X[select.relevant_features], y)
```

```python
from sklearn.pipeline import make_pipeline

model = make_pipeline(FeatureSelector(), RandomForestClassifier())
```

```python
X.shape
```

```python
from sklearn.model_selection import RepeatedKFold

kf = RepeatedKFold(n_splits=10, n_repeats=10)

predictions = list()
truth = list()
for train, test in kf.split(X):
    model.fit(X.iloc[train], y.iloc[train])
    predictions.append(model.predict_proba(X.iloc[test]))
    truth.append(y.iloc[test])
```

```python
p = np.concatenate(predictions)
t = np.concatenate(truth)
p, t
```

```python
# As shown above, truths correspond with the 2nd column of predictions. 
# The predictions are also quite confident, likely due to the ease of separtion between classes for the features
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(t, p[:,1])
         
```

```python
plt.plot(fpr, tpr)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(t, p[:,1])
```

It's likely this score would decrease given more data, or if the data collected was more varied (e.g. changing to shorter/ longer strides for running and walking)
